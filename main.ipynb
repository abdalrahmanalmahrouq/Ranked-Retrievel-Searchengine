```python
import os
import pandas as pd
import numpy as np
import math
```

```python
# To Read the data from the files
collection=[]
dict=['Economy','Gaza','Sport']
for i in dict:
    for j in range(1,51): # 1-50
        file_path=f"Collection/{i}/{j} {i}.txt"
        with open (file_path,"r",encoding="utf-8") as file:
            collection.append(file.read())
# This collection consist of all document from all category (Economy and Gaza and Sport)     
       
# convert to lower case
collection = [s.lower() for s in collection]

split_words=[]
for i in collection:
    split_words.append(i.split()) # Here to convert each index of document to sperate set of words

# To Make bag of words array (contains all distinct words from the collection )
bagOfWords=[]
for doc in split_words:
    for j in range(len(doc)):
        if doc[j] in bagOfWords:
            continue
        else :
            bagOfWords.append(doc[j])

# Here to create 150 array of document each array have len(bag of words) and begins with 0 values.
# To calculate term frequancy
term_freq_document=np.zeros((150, len(bagOfWords)), dtype=int)

# Here to calculate term frequancy for each document
for word in range(len(bagOfWords)):
    for k in range(len(split_words)):
        for j in range(len(split_words[k])):
            if bagOfWords[word]== split_words[k][j]:
                term_freq_document[k][word]+=1

print(bagOfWords)
print(split_words)
print(term_freq_document)
print(term_freq_document.shape)
```

To visualize the data

```python
if term_freq_document.shape[0]==150:
    term_freq_document=term_freq_document.T
print(term_freq_document.shape)
# Here i implement transpose method , To help me build table in pandas

docs=[]
for i in range(1,151):
    docs.append(f'doc{i}') # To rename each column

```

```python

result_table=pd.DataFrame(data=term_freq_document,index=bagOfWords,columns=docs)
print(result_table)
```

```python
# Here to calculate the log term frequancy for every document 

# Here i do tanspose for the array to make it easy to calculate
# (each document represented as index of array) 
if term_freq_document.shape[0]!=150:
    term_freq_document=term_freq_document.T

log_term_frequancy=np.empty((150, len(bagOfWords)))# 2d empty array

for i in range(150):
    for j in range(len(bagOfWords)):
        if term_freq_document[i][j]>0:
            log_term_frequancy[i][j]=round(1+math.log10(term_freq_document[i][j]),2)
        else :
            log_term_frequancy[i][j]=0    
print(log_term_frequancy)
print(log_term_frequancy.shape)                     
```

```python
# To calculate the document frequancy for each word in bag of words array

# Here i return it to the previous shape for the same reason to make it easy to calculate
if term_freq_document.shape[0] == 150:
    term_freq_document=term_freq_document.T

document_frequancy=[] # initialise the vector

for i in range(len(term_freq_document)):
    sum_d=0
    for j in range(150):
        if term_freq_document[i][j]>0:
            sum_d+=1
    document_frequancy.append(sum_d) 

print(document_frequancy)
print(len(document_frequancy))
```

```python
# To calculate the idf
idf_document_frequancy=[]
for i in document_frequancy:
    idf_document_frequancy.append(round(math.log10(150/i),2))
print(idf_document_frequancy)
```

```python
# Here to compute the tf-idf 
weighted_array=np.empty((150,(log_term_frequancy.shape[1])))

# Multiply each document log term * inverse document frequancy 
for i in range (log_term_frequancy.shape[0]):
        weighted_array[i]=log_term_frequancy[i]*(idf_document_frequancy) #tf-idf

print(weighted_array)
print(weighted_array.shape)
```

### proccessing the query term 


```python
# Ask the user to enter a query 
query=input("Please enter your query .....>")
query=query.split() # convert the query to split word then to lower case 
query = [s.lower() for s in query]
print(query)
```

```python
# To prepare the query term frequancy regrading to bagofwords array
term_freq_query=np.zeros((len(bagOfWords)))
for word in range(len(bagOfWords)):
    for q in range(len(query)):
        if bagOfWords[word]==query[q]:
            term_freq_query[word]+=1
        else :
            continue 
```

```python
# Just to see the query term frequancy 
query_result=pd.DataFrame(term_freq_query,index=bagOfWords,columns=['query term frequancy'])
print(query_result)
```

```python
# Here to calculate the log term frequancy for the query array
query_log_term_frequancy=[]
for i in range(len(term_freq_query)):
    if (term_freq_query[i]>0):
        query_log_term_frequancy.append(1+math.log10(term_freq_query[i]))
    else :
        query_log_term_frequancy.append(0)    
print(query_log_term_frequancy)
print(len(query_log_term_frequancy))        
```

```python
# Here to calculate tf-idf for query array 
# Why i implement the zip array the zip array hold the first element in query log term frequancy and idf document frequancy and multiply them
# then store the result at the first index in the weighted_query array
weighted_query = [a * b for a, b in zip(query_log_term_frequancy, idf_document_frequancy)]# tf-idf
print(weighted_query)
print(len(weighted_query))
```

### Measure the similarity (cos,jaccard)

#### Measure based on cosine similarity

```python
# Here to calculate the cosine similarity((q' dotproduct d')/(|q|*|d|)) for the query with each document 
similarity=[]
# convert it to numpy array to use dot product function
weighted_query=np.array(weighted_query)
weighted_array=np.array(weighted_array)

for doc in range(weighted_array.shape[0]):
    squared_sum_query=sum(element**2 for element in weighted_query)# to calculate the squared and summation for the vector to implement sqrt method in next line code
    squared_sum_doc  =sum(element**2 for element in weighted_array[doc])
    calculate_cos_sim=round((weighted_query.dot(weighted_array[doc]))/(math.sqrt(squared_sum_query)*math.sqrt(squared_sum_doc)),2) # implement the cosine equation
    similarity.append(calculate_cos_sim)
print(similarity)
print(len(similarity))

orderTheSimilarity=pd.DataFrame(data=similarity,index=docs,columns=['similarity'])
orderTheSimilarity=orderTheSimilarity.sort_values(by='similarity',ascending=False)
print(orderTheSimilarity)
```

#### Measure based on jaccard

```python
# Here to calculate the jaccard similarity ((q' intersect d')/(q' union d'))
query_array_jaccard=np.array(term_freq_query,dtype=int)  
for i in range(len(query_array_jaccard)):# Here to convert the query frequancy array to 0 1 array  because jaccard
    if (query_array_jaccard[i]>0):
        query_array_jaccard[i]=1
    else :
        continue    


# Transpose to easy calculate
if(term_freq_document.shape[1]==150):
    array_of_document_jaccard=term_freq_document.T
    array_of_document_jaccard=np.array(array_of_document_jaccard)

for i in range(array_of_document_jaccard.shape[0]):# To convert the term frequancy array to 0 1 array because jaccard
    for j in range(array_of_document_jaccard.shape[1]):
        if (array_of_document_jaccard[i][j]>0):
            array_of_document_jaccard[i][j]=1
        else :
            continue
        
# Then calculate the intersect and the union between the query and each document then store the value in jaccard similarity array        
jaccard_similarity=[]
for i in range(array_of_document_jaccard.shape[0]):
    counter_intersect=0
    counter_union=0
    for a,b in zip(query_array_jaccard,array_of_document_jaccard[i]):
        if (a==1 and b==1):
            counter_intersect+=1
            counter_union+=1
        elif(a==1 and b==0):
            counter_union+=0  
        elif(a==0 and b==1):
            counter_union+=1
        else :
            continue
         
    calculate_jaccard_sim=counter_intersect/counter_union
    jaccard_similarity.append(calculate_jaccard_sim)          

orderJaccardSimilarity=pd.DataFrame(data=jaccard_similarity,index=docs,columns=['similarity'])
orderJaccardSimilarity=orderJaccardSimilarity.sort_values(by='similarity',ascending=False)
print(orderJaccardSimilarity)
```
